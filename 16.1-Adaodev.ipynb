```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df=pd.read_csv('dataset/16-diabetes.csv')
import warnings
warnings.filterwarnings("ignore")
```

```python
X=df.drop("Outcome",axis=1)
y=df["Outcome"]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=15)
```

```python
col_to_fill = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in col_to_fill:
    median_value=X_train[X_train[col] != 0][col].median()
    X_train[col] = X_train[col].replace(0, median_value)
    X_test[col] = X_test[col].replace(0, median_value) 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada=AdaBoostClassifier()
ada.fit(X_train_scaled,y_train)
y_pred=ada.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
rf=RandomForestClassifier()
knn=KNeighborsClassifier()
lr=LogisticRegression()
```

```python
X=df.drop("Outcome",axis=1)
y=df["Outcome"]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=15)
col_to_fill = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in col_to_fill:
    median_value=X_train[X_train[col] != 0][col].median()
    X_train[col] = X_train[col].replace(0, median_value)
    X_test[col] = X_test[col].replace(0, median_value)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# her modelin hiperparametre olmadan sonucu
for model in [rf, knn, lr]:
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    print(f"Accuracy for {model}: {accuracy_score(y_test, y_pred)}")
    print(f"Confusion Matrix for {model}:\n", confusion_matrix(y_test, y_pred))
    print(f"Classification Report for {model}:\n", classification_report(y_test, y_pred))
    print("\n")
```

```python
# her model için parametre aralığını tanımla
param_grid_all = {
    'ada': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 1, 5, 10],
        'estimator': [None, 'DecisionTreeClassifier', 'RandomForestClassifier'],
        'algorithm': ['SAMME', 'SAMME.R']
    },
    'rf': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10],
        'criterion': ['gini', 'entropy'],
        'max_features': ['auto', 'sqrt', 'log2']
    },
    'knn': {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance'],
        'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
    },
    'lr': {
        'C': [0.01, 0.1, 1],
        'solver': ['liblinear', 'saga', 'lbfgs'],
        'penalty': ['l1', 'l2', 'elasticnet', 'none']
    }
}
```

```python
# her modeli döngü içinde eğit ve en iyi parametreleri bul
for model in ['ada', 'rf', 'knn', 'lr']:
    if model == 'ada':
        grid_search = GridSearchCV(ada, param_grid_all[model], cv=5, n_jobs=-1)
    if model == 'rf':
        grid_search = GridSearchCV(rf, param_grid_all[model], cv=5, n_jobs=-1)
    elif model == 'knn':
        grid_search = GridSearchCV(knn, param_grid_all[model], cv=5, n_jobs=-1)
    elif model == 'lr':
        grid_search = GridSearchCV(lr, param_grid_all[model], cv=5, n_jobs=-1)

    grid_search.fit(X_train_scaled, y_train)
    print(f"Best parameters for {model}: {grid_search.best_params_}")
    print(f"Best score for {model}: {grid_search.best_score_:.4f}")
    y_pred = grid_search.predict(X_test_scaled)
    print(f"Test Accuracy for {model}: {accuracy_score(y_test, y_pred):.4f}")
    print(f"Confusion Matrix for {model}:\n{confusion_matrix(y_test, y_pred)}")
    print(f"Classification Report for {model}:\n{classification_report(y_test, y_pred)}\n")
```
